```python
#functio to detect outliers
```

```python
def detect_out(columns):
    outlierindices=[]
    for column in columns:
        q1=stgrades[column].quantile(0.25)
        q3=stgrades[column].quantile(0.75)
        iqr=q3-q1
        mask=(stgrades[column]>=(q1-1.5*iqr))&(stgrades[column]<=(q3+1.5*iqr))
        mask=mask.to_numpy()
        falseindices=np.argwhere(~mask)
        outlierindices.append(falseindices)
    return np.unique(np.concatenate(outlierindices).ravel())
```

```python
num_cols = ["age", "absences"]
detect_out(num_cols)
```
