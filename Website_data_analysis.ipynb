```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/TheiScale/YouTube-Video-Notes/refs/heads/main/Website%20Data%20Analysis%20Project/data-export%20(1).csv")
```

```python
df.head()
```

```python
df.columns=df.iloc[0]
df=df.drop(index=0).reset_index(drop=True)
df.columns=["channel group","DateHour","Users","Sessions","Engaged Sessions","Average engagement time per session","Engaged sessions per user","Events per session","Engagement rate","Event count"]
```

```python
df.head()
```

```python
df['DateHour']=pd.to_datetime(df['DateHour'],format='%Y%m%d%H',errors='coerce')
df.info()
```

```python
df.head(2)
```

```python
numeric_cols=df.columns[2:]
df[numeric_cols]=df[numeric_cols].apply(pd.to_numeric,errors='coerce')
df['Hour']=df['DateHour'].dt.hour
```

```python
df.info()
```

```python
df.head()
```

```python
df.describe()
```

# Performing EDA

### Sessisons and Users  over time

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,5))
df.groupby("DateHour")[["Sessions","Users"]].sum().plot(ax=plt.gca())
plt.title("Sessions and Users over time")
plt.ylabel("Count")
plt.xlabel("DateHour")
plt.show()
```

### Total Users by Channel

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df,x="channel group",y="Users",estimator=np.sum,palette="viridis")
plt.title("Total Users by Channel")
plt.xticks(rotation=45)
plt.show()
```

### Average Engagement Time by Channel

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df,x="channel group",y="Average engagement time per session",estimator=np.mean,palette="Set2")
plt.title("Average Engagement Time by Channel")
plt.xticks(rotation=45)
plt.show()
```

## Engagement Rate Distribution by Channel

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df,x="channel group",y="Engagement rate",palette="Set2")
plt.title("Engagement Rate Distribution by Channel")
plt.xticks(rotation=45)
plt.show()
```

### Engaged vs Non Engaged Sessions

```python
df.head()
```

```python
sessions_df=df.groupby("channel group")[["Sessions","Engaged Sessions"]].sum().reset_index()
sessions_df['Non Engaged Sessions']=sessions_df['Sessions']-sessions_df['Engaged Sessions']
sessions_df_melted=sessions_df.melt(id_vars="channel group",value_vars=["Engaged Sessions","Non Engaged Sessions"])
plt.figure(figsize=(10,5))
sns.barplot(data=sessions_df_melted,x="channel group",y="value",hue="variable")
plt.title("Engaged vs Non Engaged Sessions")
plt.xticks(rotation=45)
plt.show()
```

### Traffic by Hour and Channel

```python
heatmap_data=df.groupby(["Hour","channel group"])["Sessions"].sum().unstack().fillna(0)
plt.figure(figsize=(10,5))
sns.heatmap(heatmap_data,annot=True,cmap="YlGnBu",fmt='.0f')
plt.title("Traffic by Hour and Channel")
plt.xlabel("Channel Group")
plt.ylabel("Hour")
plt.show()
```

### Engagement rate vs Sessions over time

```python
df_plot=df.groupby("DateHour")[["Engagement rate","Sessions"]].mean().reset_index()
plt.figure(figsize=(10,5))
sns.lineplot(data=df_plot,x="DateHour",y="Engagement rate",label="Engagement Rate")
sns.lineplot(data=df_plot,x="DateHour",y="Sessions",label="Sessions")
plt.xlabel("DateHour")
plt.ylabel("Count")
plt.title("Engagement rate vs Sessions over time")
plt.legend()
plt.show()
```
